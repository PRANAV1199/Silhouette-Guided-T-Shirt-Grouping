**K-Mean Algorithm**

```python
#K-Means Algorithm
def K_MEANS(X, k):
    bool_val = False
    Y = range(len(X))
    C = np.random.choice(Y, size=k, replace=bool_val)
    data = X
    ind=0
    C = X[C,:]
    data_type = int
    clusters = np.zeros(len(X), dtype=data_type)
    for i in range(data.shape[ind]):
        for j in range(i+1,data.shape[ind]):
            np.mean(data[j])
    iter = 0
   
    while iter < 20:
        i = 0
        while i < len(X):
            max_dist = 123123123.0
            j = 0
            while j < k:
                try:
                    Dist_Point_Cluster = Distance_Between_Points(C[ind+j],X[ind+i])
                    if(Dist_Point_Cluster < max_dist):
                        max_dist = Dist_Point_Cluster
                        clusters[ind+i]=j
                except:
                    print("An error occurred while computing the distance.")
                    break
                j += 1
            i += 1

        iter += 1   
        groups = [[] for _ in range(k)]
        Cluster_Circle = [[] for _ in range(k)]

        Z = range(len(X))
        for i in Z:
            groups[clusters[ind+i]].append(X[ind+i])
            Cluster_Circle[clusters[ind+i]].append(ind+i)
        T = range(k)
        for i in T:
            if len(groups[ind+i]) > ind+1:
              C[i]=list(np.mean(groups[ind+i],axis=ind))
            else :
              C[i]=list(groups[ind+i])

    return [C, groups, Cluster_Circle, clusters]
```

```python

def Distance_Betweeen(cluster1, cluster2):
    min_dist = float('inf')
    i = 0
    while i < len(cluster1):
        j = 0
        while j < len(cluster2):
            try:
                dist_between_Xi_and_Xj = Distance_Between_Points(X[cluster1[i]], X[cluster2[j]])
                if dist_between_Xi_and_Xj < min_dist:
                    min_dist = dist_between_Xi_and_Xj
                j += 1
            except IndexError:
                print("Error: Index out of range")
                break
        i += 1
    return min_dist
```

**Silhouette_Coeff**

```python
def Silhouette_Coeff(X, Cluster_Circle, clusters):

    scores = []
    for i in range(0, len(X)):
        B = {}
        cluster_no_of_Xi = clusters[i]
        a = 0.0
        Ni = len(Cluster_Circle[cluster_no_of_Xi])
        si = 0.0
        bag = []
        ind = 0
        Intra_Distance = 0.0
        j = 0
        Intra_Distance = 0.0
        B = {}
        while j < len(X):
            try:
                if j != i:
                    if clusters[ind+j] == cluster_no_of_Xi:
                        Intra_Distance += Distance_Between_Points(X[ind+i], X[ind+j])
                    else:
                        key = clusters[ind+j]
                        if key in B:
                            B[key] += Distance_Between_Points(X[ind+i], X[ind+j])
                        else:
                            B[key] = Distance_Between_Points(X[ind+i], X[ind+j])
                j += 1

            except Exception as e:
                print(f"Error: {e}")
                break

        B_items = B.items()
        for key, d in B_items:
            x=len(Cluster_Circle[key])
            n_key = x
            d_avg = (d/n_key)
            bag.append(d_avg)

        
        set_size = 1
        ai  = (Intra_Distance/ (Ni-set_size) )
        bi = min(bag)
            
        si = 0.0
        if Ni>set_size : 
            maximum = max(ai,bi)
            si = (bi - ai)/max(ai, bi)

        scores.append(si)
    data_type = np.float64
    mean_silhouette_coeff = np.mean(scores, dtype=data_type)

    return mean_silhouette_coeff
```
